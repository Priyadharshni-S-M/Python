### Task 4.3 - Convert the column to the correct datatype

```python
## With the nulls imputed, we need to remove the impurities 
## Let's write a custom function to do the same
```

```python
def cleaner(str):
    clean_str="".join(filter(lambda x: (x in "0123456789."), str))
    return float(clean_str)
```

**Explanation** - The function `cleaner` takes a string and filters out all the numeric digits and the dot symbol, joins them and then converts the given string to a float value

```python
df['C_Rating'] = df['C_Rating'].apply(cleaner)
```

## Use Regular Expressions!

```python
## Regular expressions can help in faster cleaning of data and process things much easily
## Explain a bit about this
import re
re.sub('[^0-9]','',"$137K-$171K (Glassdoor est.)")
```

```python
df2['Salary Estimate'] = df2['J_Salary'].apply(lambda x : re.sub('[^0-9-]','',x))
df2.head()
```

```python
df2.insert(2, 'MinSal', df2['Salary Estimate'].apply(lambda x: float(x.split('-')[0])))
df2.head()
```

```python
df2.insert(3, 'MaxSal', df2['Salary Estimate'].apply(lambda x: float(x.split('-')[1])))
df2.head()
```
